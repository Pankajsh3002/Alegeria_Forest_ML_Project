```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"D:\Python_AI\Python_Code\DataSets\Algerian_forest_fires_cleaned_dataset.csv")
```

```python
df.head()
```

```python
df.drop(['day','month','year'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Classes']=np.where(df['Classes'].str.strip()=='not fire',0,1)
```

```python
df['Classes'].value_counts()
```

```python
# plt.style.use('seaborn')
df.hist(bins=50,figsize=(20,15))
```

```python
percentage=df['Classes'].value_counts(normalize=True)*100
```

```python
classlabel=['Fire','Not Fire']
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabel,autopct='%1.1f%%')
plt.title('pie chart')
```

```python
sns.heatmap(df.corr())
```
